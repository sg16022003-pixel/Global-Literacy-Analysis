```python
import pandas as pd
import numpy as np

# Rename columns in source dataframes to match expected names for feature engineering
# Rename for df_Adult
df_Adult_renamed = df_Adult.rename(columns={
    'adult_literacy_rate__population_15plus_years__both_sexes__pct__lr_ag15t99': 'Literacy_Rate_Total'
})

# Rename for df_youth_literacyrate
df_youth_literacyrate_renamed = df_youth_literacyrate.rename(columns={
    'youth_literacy_rate__population_15_24_years__male__pct__lr_ag15t24_m': 'Literacy_Rate_Male',
    'youth_literacy_rate__population_15_24_years__female__pct__lr_ag15t24_f': 'Literacy_Rate_Female'
})

# Rename for df_gdp_per_capita
df_gdp_per_capita_renamed = df_gdp_per_capita.rename(columns={
    'ny_gdp_pcap_pp_kd': 'GDP'
})

# Rename for df_average_years_of_school
df_average_years_of_school_renamed = df_average_years_of_school.rename(columns={
    'mf_youth_and_adults__15_64_years__average_years_of_education': 'Years_of_Schooling',
    'literacy_rate': 'Literacy_Rate_from_Avg_School'
})

# Merge the dataframes into a single 'df'
# Start with GDP and Average Years of Schooling, also containing an existing literacy_rate column
df = pd.merge(df_gdp_per_capita_renamed, df_average_years_of_school_renamed,
              on=['entity', 'code', 'year'], how='outer', suffixes=('_gdp', '_avg_school'))

# Merge with Adult Literacy Rate (Literacy_Rate_Total)
df = pd.merge(df, df_Adult_renamed,
              on=['entity', 'code', 'year'], how='outer', suffixes=('_merged', '_adult'))

# Merge with Youth Literacy Rates (Literacy_Rate_Male, Literacy_Rate_Female)
df = pd.merge(df, df_youth_literacyrate_renamed,
              on=['entity', 'code', 'year'], how='outer', suffixes=('_merged', '_youth'))

# Post-merge cleanup and final column renaming for consistency
df = df.rename(columns={'entity': 'Country', 'year': 'Year'})

# Consolidate literacy rate into a single 'Literacy_Rate' column if multiple exist
# Prioritize 'Literacy_Rate_Total' from df_Adult, otherwise use 'Literacy_Rate_from_Avg_School'
df['Literacy_Rate'] = df['Literacy_Rate_Total'].fillna(df['Literacy_Rate_from_Avg_School'])

# Ensure data is sorted by Country and Year for the Growth Rate calculation
df = df.sort_values(by=['Country', 'Year'])

# 2. ILLITERACY %
# Calculated as 100 minus the Total Literacy Rate
df['Illiteracy_Pct'] = 100 - df['Literacy_Rate']

# 3. LITERACY GENDER GAP
# Difference between Male and Female Literacy.
# Positive = Male higher, Negative = Female higher
df['Literacy_Gender_Gap'] = df['Literacy_Rate_Male'] - df['Literacy_Rate_Female']

# 4. GDP PER SCHOOLING YEAR
# Economic output divided by average years of education
# We use .replace(0, np.nan) to avoid "Division by Zero" errors
df['GDP_per_Schooling_Year'] = df['GDP'] / df['Years_of_Schooling'].replace(0, np.nan)

# 5. EDUCATION INDEX (Normalization)
# Scales schooling years to a 0-1 range (assuming 18 years is a 'max' education)
df['Education_Index'] = df['Years_of_Schooling'] / 18

# 6. YOUTH LITERACY AVERAGE
# If you have separate male/female youth literacy, average them;
# otherwise, ensure the youth column is clean.
df['Youth_Literacy_Avg'] = (df['Literacy_Rate_Male'] + df['Literacy_Rate_Female']) / 2

# 7. LITERACY GROWTH RATE
# The percentage change in literacy compared to the previous year for THAT country
df['Literacy_Growth_Rate'] = df.groupby('Country')['Literacy_Rate'].pct_change(fill_method=None) * 100

# --- DATA CLEANUP ---
# Merges and growth calculations often create NaNs or Infinity values.
# Let's clean them up so they don't break your future charts.
df.replace([np.inf, -np.inf], np.nan, inplace=True)
# Fill NaN values in the new columns, especially where calculations might result in them (e.g., initial pct_change)
df[['Illiteracy_Pct', 'Literacy_Gender_Gap', 'GDP_per_Schooling_Year', 'Education_Index', 'Youth_Literacy_Avg', 'Literacy_Growth_Rate']] = \
    df[['Illiteracy_Pct', 'Literacy_Gender_Gap', 'GDP_per_Schooling_Year', 'Education_Index', 'Youth_Literacy_Avg', 'Literacy_Growth_Rate']].fillna(0)


print("Feature Engineering Complete!")
print(f"New Columns Added: {list(df.columns[-7:])}")
display(df.head())
```

Join Queries

```python
query_join = """
SELECT
    econ.country_name,
    econ.gdp_per_capita,
    edu.years_of_schooling
FROM Df_gdp_schooling AS econ
JOIN Df_gdp_schooling AS edu
    ON econ.country_name = edu.country_name AND econ.year = edu.year
WHERE econ.year = 2020
  AND edu.years_of_schooling < 6
ORDER BY econ.gdp_per_capita DESC
LIMIT 10;
"""

joined_data = pd.read_sql_query(query_join, conn)
print(joined_data)
```

```python
query_join2 = """
SELECT
    s.country_name,
    s.years_of_schooling,
    l.literacy_rate,
    (100 - l.literacy_rate) AS illiteracy_rate
FROM Df_gdp_schooling AS s
JOIN Df_gdp_schooling AS l
    ON s.country_name = l.country_name AND s.year = l.year
WHERE s.years_of_schooling > 10
  AND l.literacy_rate < 95  -- Identifying countries with a noticeable illiteracy gap
ORDER BY l.literacy_rate ASC
LIMIT 10;
"""

joined_data2 = pd.read_sql_query(query_join2, conn)
print(joined_data2)
```

```python
query_join3 = """
SELECT
    g.year,
    g.gdp_per_capita,
    l.literacy_rate
FROM Df_gdp_schooling AS g
JOIN Df_literacy AS l
    ON g.country_name = l.country_name
    AND g.year = l.year
WHERE g.country_name = 'Vietnam'
  AND g.year BETWEEN 2004 AND 2024
ORDER BY g.year ASC
LIMIT 10;
"""

joined_data3 = pd.read_sql_query(query_join3, conn)
print(joined_data3)
```

```python
query_join4 = """
SELECT
    e.country_name,
    e.gdp_per_capita,
    l.literacy_gender_gap AS gender_gap
FROM Df_gdp_schooling AS e
JOIN Df_literacy AS l
    ON e.country_name = l.country_name
    AND e.year = l.year
WHERE e.year = 2020
  AND e.gdp_per_capita > 30000
ORDER BY ABS(l.literacy_gender_gap) DESC
LIMIT 10;
"""

joined_data4 = pd.read_sql_query(query_join4, conn)
print(joined_data4)
```

```python
import streamlit as st
import pandas as pd
import sqlite3
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
st

# --- DATABASE CONNECTION HELPER ---
def run_query(query):
    # Ensure your database file (e.g., 'data.db') is in the same folder
    with sqlite3.connect('data.db') as conn:
        return pd.read_sql_query(query, conn)

# --- PAGE 1: SQL QUERY EXECUTOR ---
def query_executor_page():
    st.title("🔍 SQL Query Executor")
    st.markdown("Enter your SQL query below to fetch data and visualize it.")

    query = st.text_area("SQL Query", "SELECT * FROM indicators LIMIT 10")

    if st.button("Run Query"):
        try:
            df = run_query(query)
            st.success("Query successful!")

            st.subheader("Data Preview")
            st.dataframe(df)

            # Simple automatic chart if numeric columns exist
            numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
            if len(numeric_cols) >= 1:
                st.subheader("Quick Visualization")
                fig = px.bar(df, x=df.columns[0], y=numeric_cols[0], title="Results Plot")
                st.plotly_chart(fig)
        except Exception as e:
            st.error(f"Error: {e}")

# --- PAGE 2: EDA VISUALIZATIONS ---
def eda_page():
    st.title("📊 EDA Visualizations")
    df = run_query("SELECT * FROM indicators") # Load full dataset for EDA

    col1, col2 = st.columns(2)

    with col1:
        st.subheader("Distribution (Seaborn)")
        fig, ax = plt.subplots()
        sns.histplot(df.iloc[:, -1], kde=True, ax=ax) # Assuming last column is numeric
        st.pyplot(fig)

    with col2:
        st.subheader("Correlation Matrix (Heatmap)")
        fig, ax = plt.subplots()
        sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', ax=ax)
        st.pyplot(fig)

# --- PAGE 3: COUNTRY PROFILE PAGE ---
def country_profile_page():
    st.title("🌍 Country Profile")

    # Fetch unique countries for the dropdown
    countries = run_query("SELECT DISTINCT CountryName FROM indicators")['CountryName'].tolist()
    selected_country = st.selectbox("Select a Country", countries)

    if selected_country:
        df_country = run_query(f"SELECT * FROM indicators WHERE CountryName = '{selected_country}'")

        st.subheader(f"Indicators for {selected_country}")
        st.dataframe(df_country)

        # Line chart over time
        if 'Year' in df_country.columns:
            # Assuming there's a Value column
            fig = px.line(df_country, x='Year', y=df_country.columns[-1],
                          title=f"Trends over Time for {selected_country}")
            st.plotly_chart(fig)

# --- NAVIGATION SETUP ---
pg = st.navigation([
    st.Page(query_executor_page, title="Query Executor", icon="⌨️"),
    st.Page(eda_page, title="EDA Visualizations", icon="📈"),
    st.Page(country_profile_page, title="Country Profile", icon="🌍"),
])

st.set_page_config(page_title="Data Dashboard", layout="wide")
pg.run()
```

```python
import streamlit as st
import sqlite3
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. DATA ACCESS & CLEANING ---
@st.cache_data # This prevents the app from re-running cleaning on every click
def load_and_clean_data():
    # Connect to SQLite - Corrected database name to 'my_database.db'
    conn = sqlite3.connect('my_database.db')
    # Corrected SQL query to join the three tables
    df = pd.read_sql_query(
        """
        SELECT
            gdp.country_name,
            gdp.year,
            gdp.gdp_per_capita,
            gdp.years_of_schooling,
            literacy.literacy_rate AS adult_literacy_rate,
            literacy.literacy_gender_gap,
            illiteracy.illiteracy_rate AS illiteracy_pct,
            illiteracy.literacy_rate AS illiteracy_literacy_rate,
            gdp.owidregion AS region_gdp
        FROM Df_gdp_schooling AS gdp
        LEFT JOIN Df_literacy AS literacy
            ON gdp.country_name = literacy.country_name AND gdp.year = literacy.year
        LEFT JOIN Df_Illiteracy AS illiteracy
            ON gdp.country_name = illiteracy.country_name AND gdp.year = illiteracy.year
        """, conn)
    conn.close()

    # Explicitly convert numerical columns to float, coercing errors
    numeric_cols = [
        'gdp_per_capita', 'years_of_schooling', 'adult_literacy_rate',
        'literacy_gender_gap', 'illiteracy_pct', 'illiteracy_literacy_rate'
    ]
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    # Cleaning Steps (apply to the merged DataFrame)
    df.drop_duplicates(inplace=True)
    for col in df.columns:
        if df[col].dtype == 'object':
            # Fill any NaN in object columns with 'Unknown' first
            df[col] = df[col].fillna('Unknown')
            # Then explicitly ensure it's a string type
            df[col] = df[col].astype(str)
        else:
            # For numerical columns (now properly coerced), fill with median
            df[col] = df[col].fillna(df[col].median())

    return df

# --- 2. THE APP INTERFACE ---
st.set_page_config(page_title="Data Science Workflow", layout="wide")
df = load_and_clean_data()

tabs = st.tabs(["🧹 Data Cleaning", "📊 Univariate", "🔗 Bivariate"])

# --- TAB 1: DATA CLEANING PREVIEW ---
with tabs[0]:
    st.header("Cleaned Data Snapshot")
    st.write(f"Rows: {df.shape[0]} | Columns: {df.shape[1]}")
    st.dataframe(df.head(10))

# --- TAB 2: UNIVARIATE ANALYSIS ---
with tabs[1]:
    st.header("Univariate Analysis (Single Variable)")
    column = st.selectbox("Select Column to Analyze", df.columns)

    col1, col2 = st.columns(2)
    with col1:
        if df[column].dtype == 'object' or isinstance(df[column].dtype, pd.StringDtype):
            # Categorical: Count Plot
            fig = px.histogram(df, x=column, title=f"Distribution of {column}")
            st.plotly_chart(fig)
        else:
            # Numerical: Histogram/Boxplot
            fig = px.box(df, y=column, title=f"Boxplot of {column}")
            st.plotly_chart(fig)
    with col2:
        st.write(f"**Summary Statistics for {column}:**")
        st.write(df[column].describe())

# --- TAB 3: BIVARIATE ANALYSIS ---
with tabs[2]:
    st.header("Bivariate Analysis (Two Variables)")

    c1, c2 = st.columns(2)
    with c1:
        var_x = st.selectbox("X Axis", df.columns, index=0)
    with c2:
        var_y = st.selectbox("Y Axis", df.columns, index=1)

    st.subheader(f"Relationship: {var_x} vs {var_y}")

    # Logic to choose the right plot
    # Only allow scatter plot if both are numeric or can be coerced
    if (df[var_x].dtype != 'object' and not isinstance(df[var_x].dtype, pd.StringDtype)) and \
       (df[var_y].dtype != 'object' and not isinstance(df[var_y].dtype, pd.StringDtype)):
        fig = px.scatter(df, x=var_x, y=var_y, trendline="ols")
    elif (df[var_x].dtype == 'object' or isinstance(df[var_x].dtype, pd.StringDtype)) and \
         (df[var_y].dtype != 'object' and not isinstance(df[var_y].dtype, pd.StringDtype)):
        fig = px.box(df, x=var_x, y=var_y)
    else:
        st.warning("Cannot plot selected combination. Please choose appropriate numerical/categorical variables.")
        fig = px.scatter(title="Unsupported Plot Type") # Placeholder plot

    # Fixed use_container_width deprecation
    st.plotly_chart(fig, width='stretch')
```

```python
%%writefile app.py
import streamlit as st
import sqlite3
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. DATA ENGINE (Connection & Cleaning) ---
@st.cache_data
def get_cleaned_data():
    # 1. Connect to SQLite
    conn = sqlite3.connect('my_database.db')
    df = pd.read_sql_query("SELECT * FROM indicators", conn)
    conn.close()

    # 2. Data Cleaning Pipeline
    df.drop_duplicates(inplace=True)
    # Convert Year to numeric if it isn't
    if 'Year' in df.columns:
        df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
    # Fill missing values for numeric columns
    numeric_cols = df.select_dtypes(include=['number']).columns
    df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

    return df

# Helper for SQL Executor
def run_custom_query(query):
    with sqlite3.connect('my_database.db') as conn:
        return pd.read_sql_query(query, conn)

# --- 2. PAGE DEFINITIONS ---

def sql_executor_page():
    st.title("🔍 SQL Query Executor")
    query = st.text_area("Enter your SQL Query:", "SELECT * FROM indicators LIMIT 10")

    if st.button("Execute"):
        try:
            res_df = run_custom_query(query)
            st.dataframe(res_df)

            # Auto-charting for query results
            num_cols = res_df.select_dtypes(include=['number']).columns
            if len(num_cols) > 0:
                fig = px.bar(res_df, y=num_cols[0], title="Query Result Visualization")
                st.plotly_chart(fig)
        except Exception as e:
            st.error(f"Error: {e}")

def eda_page():
    st.title("📊 EDA: Univariate & Bivariate")
    df = get_cleaned_data()

    analysis_type = st.radio("Select Analysis Level", ["Univariate", "Bivariate"])

    if analysis_type == "Univariate":
        col = st.selectbox("Select Variable", df.columns)
        if df[col].dtype == 'object':
            fig = px.pie(df, names=col, title=f"Distribution of {col}")
        else:
            fig = px.histogram(df, x=col, marginal="box", title=f"Analysis of {col}")
        st.plotly_chart(fig)

    else: # Bivariate
        col_x = st.selectbox("X Axis", df.columns)
        col_y = st.selectbox("Y Axis", df.columns)
        color_var = st.selectbox("Color By (Optional)", [None] + list(df.columns))

        fig = px.scatter(df, x=col_x, y=col_y, color=color_var, title=f"{col_x} vs {col_y}")
        st.plotly_chart(fig)

def country_profile_page():
    st.title("🌍 Country Profile")
    df = get_cleaned_data()

    countries = df['CountryName'].unique()
    selected_country = st.selectbox("Pick a Country", countries)

    country_df = df[df['CountryName'] == selected_country]

    # Show indicators over time
    metrics = country_df.select_dtypes(include=['number']).columns.drop('Year', errors='ignore')
    selected_metric = st.selectbox("Select Indicator", metrics)

    fig = px.line(country_df, x='Year', y=selected_metric, markers=True,
                  title=f"{selected_metric} Trend for {selected_country}")
    st.plotly_chart(fig)

# --- 3. NAVIGATION ---
pg = st.navigation([
    st.Page(sql_executor_page, title="SQL Executor", icon="💾"),
    st.Page(eda_page, title="EDA Analysis", icon="📈"),
    st.Page(country_profile_page, title="Country Profile", icon="🌍"),
])

st.set_page_config(page_title="Data Portal", layout="wide")
pg.run()
```
